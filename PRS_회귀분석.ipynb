```python
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.diagnostic import het_white
from statsmodels.tsa.stattools import adfuller
from scipy.stats import shapiro
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


# 선택한 변수들로 데이터프레임 구성
selected_variables = ['sentiment', 'Economic Risk Rating', 'Financial Risk Rating',
                       'Investment Profile (C)', 'Risk for Budget Balance',
                       'Risk for Current Account as % of XGS', 'Risk for Debt Service',
                       'Risk for Foreign Debt', 'Risk for Inflation']
results_df = []
# 결과를 저장할 데이터프레임 생성
results_df = pd.DataFrame(columns=['Dependent Variable', 'R-squared', 'Normality p-value', 
                                   'Independence p-value', 'Homoscedasticity p-value', 'ADF p-value'])

# 변수별로 회귀분석 및 테스트 수행
for var in selected_variables[1:]:
    X = sm.add_constant(df[var])
    y = df['sentiment']
    model = sm.OLS(y, X).fit()

    # 정규성 검정
    residuals = model.resid
    _, normality_pvalue = shapiro(residuals)

    # 오차항의 독립성 확인
    _, independence_pvalue, _, _ = het_white(residuals, X)

    # 등분산성 확인
    _, homoscedasticity_pvalue, _, _ = het_white(residuals, X)

    # ADF 테스트 (단위근 검정)
    adf_result = adfuller(residuals)

    # 결과를 데이터프레임에 추가
    results_df = pd.concat([results_df, pd.DataFrame([{'Dependent Variable': var, 
                                                        'R-squared': round(model.rsquared, 4),
                                                        'Normality p-value': round(normality_pvalue, 4),
                                                        'Independence p-value': round(independence_pvalue, 4),
                                                        'Homoscedasticity p-value': round(homoscedasticity_pvalue, 4),
                                                        'ADF p-value': round(adf_result[1], 4),
                                                        'Coefficient': round(model.params[var], 4),
                                                        'Coefficient p-value': round(model.pvalues[var], 4)}])])


# 결과를 출력
results_df
```

```python
import statsmodels.api as sm
import pandas as pd

numeric_columns = df.select_dtypes(include=np.number)

# df: 데이터프레임, sentiment를 종속변수로 가정하고 나머지 변수들을 독립변수로 가정


# 선택할 독립변수 열들
independent_columns = ['Financial Risk Rating', 'Investment Profile (C)', 
                        'Risk for Budget Balance', 'Risk for Current Account as % of XGS', 
                        'Risk for Debt Service', 'Risk for Foreign Debt', 
                        'Risk for Inflation', 'anger', 'anticipation', 'disgust',
                        'fear', 'joy', 'sadness', 'surprise', 'trust', 'negative', 'positive', 'sentiment']


# 상수항(intercept) 추가
independent_variables = sm.add_constant(df[independent_columns])

# 종속변수
dependent_variable = df['Economic Risk Rating']

# 다중회귀모델 생성
model = sm.OLS(dependent_variable, independent_variables)

# 모델 피팅
results = model.fit()

# 회귀분석 결과 요약
print(results.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
# 다중공선성(VIF) 확인
vif_data = pd.DataFrame()
vif_data["Variable"] = independent_variables.columns
vif_data["VIF"] = [variance_inflation_factor(independent_variables.values, i) for i in range(independent_variables.shape[1])]

# VIF 데이터 출력
vif_data
```
